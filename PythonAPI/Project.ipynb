```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, StandardScaler, RobustScaler
import warnings
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
import pickle
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv("GameData.csv")
df = df.loc[:, ~df.columns.str.contains('^Unnamed')]
mapping = {'TRUE': 1, 'FALSE': 0}
columns_to_convert = ['has_round_started', 'is_round_over','player2_is_jumping', 'player2_is_crouching',
                        'player2_is_player_in_move','player1_is_jumping', 'player1_is_crouching', 'player1_is_player_in_move','player1_button_down', 'player1_button_right',
                        'player1_button_left', 'player1_button_select',
                        'player1_button_start', 'player1_button_Y',
                        'player1_button_B', 'player1_button_X',
                        'player1_button_A', 'player1_button_L',
                        'player1_button_R', 'player2_button_up',
                        'player2_button_down', 'player2_button_right',
                        'player2_button_left', 'player2_button_select',
                        'player2_button_start', 'player2_buttons.Y',
                        'player2_button_B', 'player2_button_X',
                        'player2_button_A', 'player2_button_L',
                        'player2_button_R']  # Replace with your column names
for column in columns_to_convert:
    df[column] = df[column].map(mapping)
```

```python
for i in df.columns:
    print(i)
    print(df[i].head())
    print(df[i].dtype,'\n')
```

```python
catagorical = []
for i in df.columns:
    if df[i].dtype == 'object' or df[i].dtype == 'bool':
        catagorical.append(i)
```

```python
catagorical
```

```python
keys = {'NOT_OVER': 0,
       'P1':1,
       'P2':2,
       True:1,
       False:0
       }
```

```python
for i in catagorical:
    df[i]=df[i].map(keys)
```

```python
df
```

```python
for i in df.columns:
    if 'button' in i:
        print(i)
```

```python
df = df.dropna()
```

```python
X = df.drop(['player1_button_up','player1_button_down','player1_button_right','player1_button_left','player1_button_select','player1_button_start','player1_button_Y','player1_button_B','player1_button_X','player1_button_A','player1_button_L','player1_button_R'],axis=1)
Y = df[['player1_button_up','player1_button_down','player1_button_right','player1_button_left','player1_button_select','player1_button_start','player1_button_Y','player1_button_B','player1_button_X','player1_button_A','player1_button_L','player1_button_R']]
```

```python
X
```

```python
X.isna().sum()
```

```python
sc = StandardScaler()
rb = RobustScaler()
X = sc.fit_transform(X)
X = rb.fit_transform(X)
```

```python
X
```

```python
Y
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.2)
```

```python
print(y_train.shape)
```

```python
reg = LinearRegression()
reg.fit(x_train, y_train)
y_pred = reg.predict(x_test)
y_pred = np.where(y_pred > 0.5,1,0)
print(f"loss is {mean_squared_error(y_test,y_pred)}")
```

```python
with open('Model.pkl', 'wb') as f:
    pickle.dump(reg, f)
```

```python
# Save the scaler objects and their arguments using pickle
with open('Scaler.pkl', 'wb') as f:
    pickle.dump(sc, f)
np.save('X_scaled_sc.npy', X)
with open('RobustScaler.pkl', 'wb') as f:
    pickle.dump(rb, f)
np.save('X_scaled_rb.npy', X)
```
